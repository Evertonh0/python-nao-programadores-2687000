Diferente das tuplas e das listas que usavam indices numéricos. Nos dicionarios podemos personaliza-las.

Tuplas, listas e dicionarios = Estruturas de dados

```python
dados = 'Pedro', 25

dados = { 'nome':'Pedro','idade': 25  }

print(dados['nome'])
print(dados['idade'])

# O append não funciona nos dicionarios. Tampouco é necessario.
# Bastar usar a seguinte estrutura

dados['sexo'] = 'M'

print(dados['sexo'])
```

```python
filme = {'titulo':'Tenet',
         'ano': 2020,
         'diretor': 'Christopher Nolan'
        }

print(filme.values())
print(filme.keys())

print(filme.items())
```

```python
filme = {'titulo':'Tenet',
         'ano': 2020,
         'diretor': 'Christopher Nolan'
        }
for k,v in filme.items():
  print(f'O {k} é {v}')
```

```python
musica = { 'nome':'Baila', 'Artista':'Froid', 'Album':'Baila','ano':2020}

for k,v in musica.items():
  print(f'O {k}, é {v}')
```

```python
# Exercício 1: Criando um dicionário
# Crie um dicionário chamado estudante que contenha as seguintes informações:


estudante = {'nome':'João','idade':20,'curso':'engenharia' }
print(estudante.values())
print(estudante.keys())
print(estudante.items())



# Depois, exiba o conteúdo do dicionário.
```

```python
# Exercício 2: Adicionando e Atualizando Valores
#Comece com o seguinte dicionário:

aluno = {"nome": "Ana", "idade": 22}

# Adicione uma nova chave chamada curso com o valor "Química".

aluno['curso'] = 'química'

# Atualize a idade para 23.

aluno['idade'] = 23

# Exiba o dicionário atualizado.

print(aluno.keys())
print(aluno.values())
print(aluno.items())

```

```python
# Exercício 3: Acessando Valores
# Dado o dicionário:

carro = {"marca": "Toyota", "modelo": "Corolla", "ano": 2020}

# Responda

# Qual é a marca do carro?

print('A marca do carro é', carro['marca'])


# Qual é o modelo do carro?

print('Já o modelo do é',carro['modelo'])
```

```python
# Dicionário inicial
pessoa = {"nome":"João", "idade": 25}

# Adicionando várias chaves e valores
pessoa.update({'curso':'engenharia química','cidade':'Viamão'})    # {'chave' : 'valor', 'chave' : 'valor'}


# Exibindo o dicionário atualizado
print(pessoa)
```
